```python
import pandas as pd
import numpy as np
import statistics
import matplotlib.pyplot as plt
import seaborn as sns
# Dataset
water_consumption = [100, 150, 170, 180, 200]
days = [2, 4, 1, 2, 3]

df = pd.DataFrame({
    "Water_Consumption": water_consumption,
    "Days": days
})

#  Average water consumption
avg_consumption = df["Water_Consumption"].mean()

# Total water usage per day
df["Usage_per_Day"] = df["Water_Consumption"] / df["Days"]
total_usage_per_day_avg = df["Usage_per_Day"].mean()

#  Water usage increased
df["Increase"] = df["Water_Consumption"].diff()

# Sum of daily increased water
sum_increase = df["Increase"].sum()

# Shortage condition
shortage_occurs = df["Water_Consumption"].is_monotonic_increasing

print(df)
print("Average Consumption:", avg_consumption)
print("Avg Usage per Day:", total_usage_per_day_avg)
print("Sum of Increased Water:", sum_increase)
print("Shortage Occurs:", shortage_occurs)

water_consumption = [100, 150, 170, 180, 200]
days = [2, 4, 1, 2, 3]

# Average consumption
total = 0
for v in water_consumption:
    total += v
avg_consumption = total / len(water_consumption)

# Usage per day
usage_per_day = []
for i in range(len(water_consumption)):
    usage_per_day.append(water_consumption[i] / days[i])

total_usage = 0
for v in usage_per_day:
    total_usage += v
total_usage_per_day_avg = total_usage / len(usage_per_day)

# Increase
increase = [0]
for i in range(1, len(water_consumption)):
    increase.append(water_consumption[i] - water_consumption[i - 1])

sum_increase = 0
for v in increase:
    sum_increase += v

# Shortage check
shortage_occurs = True
for i in range(1, len(water_consumption)):
    if water_consumption[i] < water_consumption[i - 1]:
        shortage_occurs = False
        break
print("Average Consumption:", avg_consumption)
print("Avg Usage per Day:", total_usage_per_day_avg)
print("Sum of Increased Water:", sum_increase)
print("Shortage Occurs:", shortage_occurs)


```
